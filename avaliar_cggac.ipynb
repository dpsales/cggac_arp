```python
from openpyxl import load_workbook

xlsx_path = 'dados_cggac(1).xlsx'
try:
    wb = load_workbook(xlsx_path, data_only=True)
    ws = wb.active
    headers = [cell.value for cell in ws[1]]
    print('Colunas do XLSX (primeiras 50):')
    for i, h in enumerate(headers[:50], 1):
        if h:
            print(f'{i:2d}. {h}')
except Exception as e:
    print(f'Erro: {e}')
```
